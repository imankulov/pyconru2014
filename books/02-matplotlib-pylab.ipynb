```python
%matplotlib inline
import numpy as np
from pylab import *
```

# Основные типы графиков pyplot

```python
x = np.random.normal(size=500)
```

### Scatter Plot

```python
scatter(x, np.arange(x.size))
xlim(-6, 6)
```

### Гистограмма

```python
title(u"Гистограмма нормального распределения")
hist(x)
xlim(-6, 6)
```

### Ящик с усами

```python
boxplot(x, vert=False)
xlim(-6, 6)
```
